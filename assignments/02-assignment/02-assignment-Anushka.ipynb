4.  What is the Stable Unit Treatment Value Assumption (SUTVA), and why is it important for causal inference? Give an example of a situation where SUTVA might be violated (not included in the slides).

## SUTVA states that the potential outcomes don't vary with treatment (there is only one treatment) and that the treatment effect on on person doesn't affect another person's outcome. This is important to make sure that there is a viable case on treatmenet and effect. 

## Ex: you wanted to measure if the day of the week you take an exam affects your score (is Monday inherently a bad day?) and you split students into random groups and had each group do it on a different day. However, if people from the monday exam go and talk to the rest of the classmates about the exam contents, then those students would gain an unfair advantage due to that interference, and would probably create the false conclusion that later in the week is better for tests - but it's really just that they knew what to expect on the exam. 

5.  What is a "balance test" (or "randomisation check")? What is its purpose and its limitations?

## A balance test compares means of covariates of groups, checks for selection bias by checking how similar the groups are BEFORE treatment, and makes sure that any potential outcomes are due to differences that didn't already exist. Balance tests can only show visible differences though, not to mention that random variation is still a thing - this cam lead to errors and false conclusions in trying to narrow down groups, especially if the sample size is already small.

6. What is post-treatment bias (or overcontrol), and why is it a problem?

## post treatment bias is the act of controlling for variables affected by the treatment. This is a problem because this can lead to us making false conclusions about the treatment effect. For example, if you wanted to test if having an emotional support animal helps improve depression, you would want to control for the symptoms of the diagnosis beforehand, not stress levels after getting a pet - controlling for post treastment effect would obscure the true treatment effect

7. A study examines the impact of a new teaching method on student test scores. Students are randomly assigned to either the new method or a traditional one. However, students in the new method group are more likely to withdraw from the study. What type of selection bias is this, and how might it affect the results?

## This is attrition bias, aka when a participant drops out of the study before it finishes. This can lead to a false conclusion on treatment effect. For example, if the people in the new group drop out, we don't know how the new method actually affects test scores, especially because it would be more likely that those who drop out are not doing well. The ones who don't drop out are probably doing better, so not factoring the dropouts can lead to the biased conclusion that the new method is improving test scores.
